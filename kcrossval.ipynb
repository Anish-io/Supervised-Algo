K-cross Validation

```python
import numpy as np

from sklearn.datasets import load_digits
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline

from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
```

```python
digits = load_digits()

X = digits.data
y = digits.target

print("Dataset shape:", X.shape)
```

```python
log_model = make_pipeline(
    StandardScaler(),
    LogisticRegression(max_iter=5000)
)

log_scores = cross_val_score(log_model, X, y, cv=5)

print("Logistic Regression CV scores:", log_scores)
print("Average Accuracy:", log_scores.mean())
```

```python
svm_model = make_pipeline(
    StandardScaler(),
    SVC(kernel='rbf')
)

svm_scores = cross_val_score(svm_model, X, y, cv=5)

print("SVM CV scores:", svm_scores)
print("Average Accuracy:", svm_scores.mean())
```

```python
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)

rf_scores = cross_val_score(rf_model, X, y, cv=5)

print("Random Forest CV scores:", rf_scores)
print("Average Accuracy:", rf_scores.mean())
```
